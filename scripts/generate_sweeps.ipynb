# Generate parameter sweeps for qvarnet experiments

```python

dof_range = range(1, 110, 10)
dof = 40
n_layers = [1, 2, 3, 4, 5]
n_hiddens_per_layer = [4, 8, 16, 32, 64, 128]
# n_epochs = [1e3, 1e4, 1e5]
# batch_sizes = [1_000, 10_000, 100_000]

print("number of jobs:",len(n_layers) * len(n_hiddens_per_layer) )
    
def generate_configs():
    # for dof in dof_range:
    for n_layer in n_layers:
        for n_hidden in n_hiddens_per_layer:
            architecture = [dof] + [n_hidden] * n_layer + [1]
            file_info = f"{dof} particles {n_layer} layers {n_hidden} hiddens per layer"
            yield f"""
model.architecture={architecture}
            """, file_info
```

```python
import os


cwd = os.getcwd()
print(cwd)
if not cwd.endswith("configs"):
    os.makedirs("configs", exist_ok=True)
    os.chdir("configs")
print(os.listdir("."))
print(os.getcwd())

job_id = 0

for config, file_info in generate_configs():
    with open(f"{job_id}.txt", "w") as f:
        f.write(f"# Job info: {file_info}\n")
        f.write(config.replace(" ", ""))
    job_id += 1
```
